```python
import itertools


def valjane_kombinacije(n):
    # Generiramo listu sa 3 broja (1, 2, 3) svaki n puta
    brojevi = [1] * n + [2] * n + [3] * n

    # Generiramo sve permutacije tih brojeva
    permutacije = set(itertools.permutations(brojevi))

    # Filtriramo permutacije koje nisu valjane (ako ima iste brojeve uzastopno)
    valjane_permutacije = [
        perm for perm in permutacije
        if all(perm[i] != perm[i + 1] for i in range(len(perm) - 1))
    ]

    return valjane_permutacije


# Unos broja n
n = int(input("Unesite broj n (koliko puta svaki broj treba biti posjeÄ‡en): "))

# Generiranje svih valjanih kombinacija
kombinacije = valjane_kombinacije(n)

# Ispis rezultata
print(f"Sve valjane kombinacije za n = {n}:")
for kombinacija in kombinacije:
    print(kombinacija)

print(f"Ukupno valjanih kombinacija: {len(kombinacije)}")
```
